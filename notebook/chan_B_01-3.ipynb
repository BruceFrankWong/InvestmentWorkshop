完整的计算K线合并，并保存数据。

## 准备

### 引用模块

```python
import pandas as pd

from utility import (
    get_available_datafile_name,
    load_csv_as_dataframe,
    get_saved_filename,
    save_dataframe_to_csv
)
from InvestmentWorkshop.indicator import (
    pbx,
    merge_candlestick_dataframe,
)
```

### 设置 pandas

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
```

### 可用数据

```python
available_datafile = get_available_datafile_name(DATA_PATH, 'Minute')
for symbol, period in available_datafile:
    print(f'symbol: {symbol:>12}, period: {period}')
```

## 计算并保存

```python
for symbol, period in available_datafile:
    print(f'\n{symbol}\n')
    
    # 加载数据
    df_data: pd.DataFrame = load_csv_as_dataframe(DATA_PATH.joinpath(f'{symbol}_{period}.csv'))

    # 丢弃不需要的 column。
    columns_origin = list(df_data.columns)
    columns_needed = ['open', 'high', 'low', 'close']
    columns_drop = [x for x in columns_origin if x not in columns_needed]
    df_data.drop(columns=columns_drop, inplace=True)

    # 显示计算用数据
    df_data.info()
    print('-' * 100)
    df_data.head(5)
    
    # 计算
    df_merged: pd.DataFrame = merge_candlestick_dataframe(df_data, debug=True)
    
    # 保存
    save_dataframe_to_csv(DATA_PATH.joinpath(f'{symbol}_merged.csv'), df_merged)
    print(f'{symbol} 的计算结果已经保存。\n')
    
print('=' * 100)
print('全部计算都已完成并保存。')
```
